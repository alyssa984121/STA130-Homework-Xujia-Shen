### Q1

#### (a): 

A **Classification Decision Tree** addresses **classification problems**, which involve predicting a categorical outcome based on input features. This means the target variable can take on discrete, predefined classes, such as "yes/no," "spam/not spam," or categories like "cat/dog/horse."

Examples of real-world applications:
1. **Email Spam Detection:** Predict if an email is spam or not based on words and phrases in the email.
2. **Medical Diagnosis:** Classify if a patient has a certain disease based on test results and symptoms.
3. **Loan Approval:** Predict if a loan should be approved or denied based on applicant characteristics like income, credit score, and employment history.
4. **Customer Segmentation:** Categorize customers into groups (e.g., "high-value," "at-risk," or "new customers") based on their purchase behavior.

In contrast, **regression problems** involve predicting a continuous numeric value, such as predicting house prices, stock values, or temperatures.

#### (b): How Does a Classification Decision Tree Make Predictions?

A **Classification Decision Tree** makes predictions by splitting the dataset into subsets based on feature values at each level of the tree. These splits are designed to maximize the "purity" of the subsets concerning the target variable. The process involves the following:

1. **Root Node:** The tree starts with a root node that contains all the data. The model chooses the feature that best splits the data into groups based on the target variable.
2. **Splitting:** At each node, the tree examines a feature and creates branches for different values (or ranges) of that feature.
3. **Leaf Nodes:** When the data in a node cannot be split further, the node becomes a leaf. Each leaf node corresponds to a prediction—a class label.

**Example:**
Consider a decision tree predicting if someone buys a product:
- Start with "Age." If age < 25, go left; if age ≥ 25, go right.
- On the left branch, check "Income." If income < $50K, predict "No"; otherwise, predict "Yes."
- On the right branch, directly predict "Yes."

This sequential decision-making process resembles a flowchart, where each split is based on a condition that minimizes classification error.

**Contrast with Multiple Linear Regression:**
- In **Multiple Linear Regression**, predictions are based on a mathematical equation:

Here, (x1, x2, ..., xn) are the input features, and the coefficients βi determine the influence of each feature on the continuous output y.

- Regression assumes a linear relationship between the inputs and output. Decision Trees, on the other hand, make **non-linear predictions** by navigating through conditions.

---

### Interaction Insight:
The **sequential splits in a decision tree** can be thought of as "interactions" between features. For instance:
- If "Age" < 25 and "Income" < $50K → Predict "No."

- If "Age" < 25 but "Income" ≥ $50K → Predict "Yes."

These combinations of decisions mimic how **interactions** between variables influence outcomes in regression, but decision trees do this through branching rather than equations.

#### 1. Accuracy

Accuracy is the ratio of correctly predicted observations to the total observations. In simpler terms, it is the proportion of correct predictions made by the model out of all predictions. This metric is often used to evaluate overall model performance but works best when the data is balanced, meaning there are roughly equal numbers of positive and negative samples.

When to use:When the positive and negative classes are balanced, accuracy is a good indicator of model performance.
When the costs of false positives (predicting positive when it’s actually negative) and false negatives (predicting negative when it’s actually positive) are similar.

Example:In a voice assistant, misrecognizing a command might not be a major issue, so accuracy is a suitable measure of performance.


Limitations:In imbalanced datasets, accuracy can be misleading. For example, in spam detection, if a model always predicts “not spam,” it may still show high accuracy, even though it’s not identifying any actual spam. In this case, accuracy wouldn’t reflect the true effectiveness of the model.

#### 2. Sensitivity

Sensitivity, or recall, measures the proportion of actual positive samples that were correctly identified by the model. In other words, it tells us how good the model is at identifying all true positives. It’s particularly important in situations where false negatives (missing positive cases) are costly.

When to use:When missing positive cases (false negatives) is costly, such as in medical diagnoses.
For example, in cancer screening, failing to identify a cancer patient can have serious consequences, so a model with high sensitivity is crucial to catch as many true positive cases as possible.
Limitations:

High sensitivity can lead to an increase in false positives (incorrectly identifying negative cases as positive). However, in some scenarios, false positives might be acceptable.

Example:In breast cancer screening, missing a cancer diagnosis could delay treatment and be life-threatening. So, high sensitivity is desired to avoid missing any potential cases.

#### 3. Specificity

Specificity is the counterpart to sensitivity. It measures the proportion of actual negative samples that were correctly identified. In other words, it reflects how good the model is at identifying negative cases. This metric is useful when the cost of false positives is high.

When to use: When false positives have high costs, or when you want to minimize the incorrect identification of negative cases as positive.

Example:In newborn genetic disease screening, false positives (incorrectly diagnosing a healthy baby) could lead to unnecessary treatments and stress, so high specificity is needed to avoid misdiagnosing healthy children.

Limitations:High specificity can lead to higher false negatives (missing some true positives). However, in some cases, false negatives may be more acceptable than false positives.

#### 4. Precision

Precision measures the proportion of predicted positive cases that are actually positive. In other words, it tells us how reliable the model is when it predicts positive cases. Precision is important in situations where false positives are costly.
When to use:When false positives are costly or cause disruption.

Example:In spam email detection, if a model wrongly classifies an important email as spam (false positive), it may result in the loss of crucial information. Thus, high precision is required to ensure that emails marked as spam are truly spam.

Limitations:High precision usually leads to lower sensitivity (fewer true positives identified), but in some cases, this is acceptable.

   - A confusion matrix compares the true labels (`y_true`) and the predicted labels (`y_pred`).
   - In binary classification, it consists of four components: 
     - **TP (True Positive)**: Correctly predicted positives.
     - **TN (True Negative)**: Correctly predicted negatives.
     - **FP (False Positive)**: Incorrectly predicted positives.
     - **FN (False Negative)**: Incorrectly predicted negatives.
   - A typical confusion matrix looks like this:
     ```
                 Predicted
                 Positive   Negative
     True Positive    TP        FN
     True Negative    FP        TN
     ```

3. **What Leads to TP, TN, FP, and FN**:
   - **TP**: The model correctly predicts the positive class (e.g., predicts "hardcover" when the book is actually hardcover).
   - **TN**: The model correctly predicts the negative class (e.g., predicts "paperback" when the book is actually paperback).
   - **FP**: The model incorrectly predicts the positive class (e.g., predicts "hardcover" when the book is actually paperback).
   - **FN**: The model incorrectly predicts the negative class (e.g., predicts "paperback" when the book is actually hardcover).

4. **Order of Arguments in `confusion_matrix` Function**:
   - When using `confusion_matrix(y_true, y_pred)`, the first argument should be `y_true` (actual values), and the second argument should be `y_pred` (predicted values).

5. **Visualizations with Decimal Numbers**:
   - It's better to visualize metrics with decimal numbers (e.g., 0.123) instead of percentages (e.g., 12.3%) for better clarity.
