```python
#Q3
#4
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import plotly.graph_objects as go

# Load data
url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
pokeaman = pd.read_csv(url).fillna('None')

# Create a binary variable for "Fire" type
pokeaman['str8fyre'] = (pokeaman['Type 1'] == 'Fire').astype(int)

# Continuous variable: Attack
# Additive model
additive_model_formula = 'Attack ~ str8fyre'
additive_model = smf.ols(additive_model_formula, data=pokeaman).fit()

# Synergistic (Interaction) model
synergistic_model_formula = 'Attack ~ str8fyre * Legendary'
synergistic_model = smf.ols(synergistic_model_formula, data=pokeaman).fit()

# Simulate random noise for more realism
pokeaman['noise'] = np.random.normal(0, 10, size=len(pokeaman))
pokeaman['Attack_with_noise'] = pokeaman['Attack'] + pokeaman['noise']

# Plot Additive Model
fig_additive = go.Figure()

# Scatter plot for the original data
fig_additive.add_trace(go.Scatter(x=pokeaman['str8fyre'], y=pokeaman['Attack_with_noise'], mode='markers', name='Data'))

# Best fit line from the additive model
fig_additive.add_trace(go.Scatter(x=pokeaman['str8fyre'], y=additive_model.fittedvalues, mode='lines', name='Best Fit Line (Additive)'))

fig_additive.update_layout(title='Additive Model: Attack vs Fire Type',
                           xaxis_title="Fire Type (Binary)",
                           yaxis_title="Attack",
                           showlegend=True)

# Plot Synergistic (Interaction) Model
fig_synergistic = go.Figure()

# Scatter plot for the original data
fig_synergistic.add_trace(go.Scatter(x=pokeaman['str8fyre'], y=pokeaman['Attack_with_noise'], mode='markers', name='Data'))

# Best fit line from the interaction model
fig_synergistic.add_trace(go.Scatter(x=pokeaman['str8fyre'], y=synergistic_model.fittedvalues, mode='lines', name='Best Fit Line (Interaction)'))

fig_synergistic.update_layout(title='Synergistic Model: Attack vs Fire Type and Legendary',
                              xaxis_title="Fire Type (Binary)",
                              yaxis_title="Attack",
                              showlegend=True)

# Show plots
fig_additive.show()
fig_synergistic.show()


```
